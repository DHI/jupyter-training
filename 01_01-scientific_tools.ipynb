#### Converting Stacked tables

The XLSX file contains time series that have been exported from the FEFLOW GUI. 

It has the form of a stacked table, meanign that data of all times and observation points appear in the same table. This is a quite common of format in model outputs in general, however not really useful for processing.

Pandas provides an in-built function to tabulate the data with an individual colum per observation bore. This is done by using the `pivot`-function

Note this useful if all observation points share the same time steps (as they usually do in model output). For measurement data with usually asynchrounous data, the data should be resampled if tabulation is required.

```python
# to tabulate the data, provide the names of the coulnms that contains the index (here: X, time), the values (here: Y)
# and the columns (here: curve)
df_pivot = df.pivot(index="X", values="Y", columns="Curve")
df_pivot.index.name = "Time"
```

```python
df_pivot
```

#### Resampling

To resample a DataFrame (or DataSeries), its index field must be of type DateTime.

```python
# example: datetime objects
from datetime import datetime
from datetime import timedelta

datetime(2018, 1, 1) + timedelta(days=1.2)
```

To convert the index axis from a time to a calendar axis, we create a new column of type DateTime and set it as the new index:

```python
# create a new colum
df_pivot["Calendar"] = pd.to_datetime(df_pivot.index, unit="D", origin=datetime(2018, 1, 1))
```

```python
# set calendar as new index:
df_pivot.set_index("Calendar", inplace=True)
df_pivot
```

The actual resampling is done with the `resample` method. The first argument is the resampling rule (`D` stands for Daily values). The method returns an object, on that we need to apply a accumulation operation. In this case, we use the `mean` value of all measured heads in a day, which makes sense for head measurements.

(A different option would be `.resample("D").sum()`, for example when downsampling rainfall or pumping rates).

```python
df_pivot.resample("D").mean()
```

Because the timesteps of the model are often longer than the resampling rates, gaps (NaN values) appear in the resulting time series, which needs to be addressed.

Typical handling strategies is gap-filling (e.g. `interpolate`) or removal (`.dropna`) of the gaps:

```python
# fill the gaps by linear interpolation
df_pivot.resample("D").mean().interpolate()
```

```python
# remove gaps by dropping all rows that contain NaN values
df_pivot.resample("D").mean().dropna()
```

###   Accessing FEFLOW (Time Series)

Time Series data can be imported from FEFLOW directly through the IFM API. There is no direct support for Pandas yet, but the following snippet does the job:

```python
# first import the FEFLOW IFM. This lenghty code snippet is auto-sensing
# to most system environments. 

import sys
import os
import platform

# load IFM
ifmversion = '71'  # choose version here
arch = platform.architecture()[0][:2]
if 'FEFLOW' + ifmversion + '_ROOT' not in os.environ:
    raise EnvironmentError('FEFLOW' + ifmversion + '_ROOT Environment Variable not set!')
sys.path.append(os.environ['FEFLOW' + ifmversion + '_ROOT'] + '\\bin' + arch)
import ifm
```

```python
# No license is required to open the model read-only:
ifm.forceLicense("Viewer")
```

```python
# copy this function into your script
def getHistoryDataFrame(doc, hist_type, hist_subtype=0):
    
    import pandas as pd
    
    ifmHistoryID = {
    "HEAD": 1,  # The 'head distribution' window
    "AVER_M": 2,  # The 'average mass distribution' window
    "LOCA_M": 3,  # The 'local mass distribution' window
    "AVER_H": 4,  # The 'average heat distribution' window
    "LOCA_H": 5,  # The 'local heat distribution' window
    "FFLUX": 6,  # The 'fluid flux' window
    "MFLUX": 7,  # The 'mass flux' window
    "HFLUX": 8,  # The 'heat flux' window
    "ITERA": 10,  # Iteration window index
    "TIMES": 12,  # Time step history index
    "RESIDUAL": 13,  # Residual window index
    "VARIO": 15,  # Kriging variogram index
    "ANA": 16,  # Analytical function index
    "MULTW_FLUX": 17,  # Muli-layer well flux window index
    "BHE": 18,  # Borehole heat exchanger window index
    "IFM": 19,  # IFM based diagram window index
    "FFLUX_I": 20,  # The 'integral fluid flux' window
    "MFLUX_I": 21,  # The 'integral mass flux' window
    "HFLUX_I": 22,  # The 'integral heat flux' window
    "BHE_IO": 23}  # The Inlet/outlet bhe diagram
    
    if type(hist_type) == str:
        try:
            hist_type = ifmHistoryID[hist_type]
        except ValueError as e:
            raise ValueError("Unknown History Type "+str(hist_type))
    
    chart = doc.getHistoryValues(hist_type, hist_subtype)

    times = chart[0]
    values = chart[1]
    itemnames = chart[2] 
    
    df = pd.DataFrame(values, columns=times, index=itemnames, ).T
    df.index.name="Time"
    
    return df
```

```python
# load a model and get the hydraulic head history
doc = ifm.loadDocument("./example_data/Pit3_cal2.dac")
getHistoryDataFrame(doc, "HEAD", 0).plot()
```

###   Accessing FEFLOW

Similar to the time series, we can export geometries directly from a fem or dac. Some code snippets are required.

Let's start with the observation points of the model:


```python
from shapely.geometry import Point, Polygon

# create a list of observation points in FEFLOW model:
obs = []
for o in range(doc.getNumberOfValidObsPoints()):
    X0, Y0 = doc.getOriginX(), doc.getOriginY()
    
    label = doc.getObsLabel(o)
    x,y,z = doc.getXOfObsId(o)+X0, doc.getYOfObsId(o)+Y0, doc.getZOfObsId(o)
    print("{} ({}|{}|{})".format(label,x,y,z))
    
    shape = Point(x,y)  # create a geometry object
    
    obs.append([label, x,y,z, shape])
```

```python
gdf_obs = gpd.GeoDataFrame(obs, columns=["label", "x", "y", "z", "shape"])
gdf_obs.set_index("label", inplace=True)
gdf_obs
```

```python
gdf_obs.set_geometry("shape", inplace=True)
gdf_obs.plot()
```

```python
# bonus: with labels
import matplotlib.pyplot as plt
gdf_obs.plot()
for label, row in gdf_obs.iterrows():
    plt.annotate(s=label, 
                 xy=(row['x'], row['y']),
                 horizontalalignment='center')
```

The numerical mesh can be shown in a similar way:

```python
"""
load the nodes coordinates, the incidence matrix and the head field
"""

def read_feflowmesh(pdoc, global_cos=True):
    
    import numpy as np
    
    dd = doc.getNumberOfDimensions()
    if global_cos:
        X0 = doc.getOriginX()
        Y0 = doc.getOriginY()
    else:
        X0 = Y0 = 0

    x = []
    y = []

    if dd == 2:
        nn = doc.getNumberOfNodes()
        ee = doc.getNumberOfElements()
        stop = 1
    else:
        nn = doc.getNumberOfNodesPerSlice()
        ee = doc.getNumberOfElementsPerLayer()
        stop = 2

    for n in range(nn):
        x.append(doc.getX(n)+X0)
        y.append(doc.getY(n)+Y0)

    x = np.array(x)
    y = np.array(y)

    imat=[]

    for e in range(ee):  # PerLayer
        if not doc.getMatElementActive(e):
            continue
        NN = doc.getNumberOfElementNodes(e) / stop
        element_nodes = [doc.getNode(e, N) for N in range(NN)]
        if NN == 3:
            imat.append(element_nodes)
        elif NN == 4:
            imat.append(element_nodes[:3])  # split quadrangle in 2 triangles
            imat.append(element_nodes[1:])
        else:
            raise ValueError(str(NN*2)+"-noded element not supported")

    return (x,y,imat)

x, y, imat = read_feflowmesh(doc)
print("Loaded {} nodes and {} elements".format(len(x),len(imat)))
```

```python
# create a GeoDataFrame from the mesh and save as shape file
gdf_elements = gpd.GeoDataFrame([Polygon([(x[n], y[n]) for n in element]) for element in imat])
gdf_elements.columns = ["element_shape"]
gdf_elements.set_geometry("element_shape", inplace=True)
gdf_elements.index.name = "ELEMENT"
gdf_elements["ELEMENT"] = gdf_elements.index.values
gdf_elements["AREA"] = gdf_elements.geometry.area

gdf_elements.plot()
```

```python
fig, ax = plt.subplots(figsize=(12,6))
plt.axis('equal')
gdf_elements.plot(ax=ax)
gdf_isolines.plot(ax=ax, column="F")
gdf_obs.plot(ax=ax, color="red")
```

### Dissolve

Dissolve merges features of the same kind (given by a column name), into one feature. This is useful to reduce the number of features, e.g. by finite-elements into parameter zones.

To combine all features in a GeoDataFrame, first set a dummy attribute and dissolve by it.

```python
gdf_elements["dummy"] = 0
gdf_elements.dissolve(by="dummy").plot()
```

###  Creating Animations

Animations are an excellent way to screen the results of transient models. 
This example creates an animation of the transient water development of a FEFLOW model.
(A similar movie could be exported from within FEFLOW using the autopilot panel).

We load the model first and plot an example figure.

```python
# load FEFLOW model
try:
    doc = ifm.loadDocument("./example_data/Pit3_cal2-highres.dac")
except StandardError as e:
    print("/example_data/Pit3_cal2-highres.dac not found!")
    print("run ./example_data/Pit3_cal2-highres.fem first to create it!")
```

```python
import numpy as np

# read node coordinates and incidence matric
x, y, imat = read_feflowmesh(doc)

# Get nodal head values from FEFLOW
def get_nodal_head_values(pdoc):
    if doc.getNumberOfDimensions() == 2:
        nn = doc.getNumberOfNodes()
    else:
        nn = doc.getNumberOfNodesPerSlice()
        
    h = [doc.getResultsFlowHeadValue(n) for n in range(nn)]   
    h = np.array(h)
    
    #replace nan heads (inactive nodes) with mean value
    inds = np.where(np.isnan(h))
    h_mean = h[~np.isnan(h)].mean()
    for i in inds:
        h[i] = h_mean
    
    return h

h = get_nodal_head_values(doc)
```

```python
# do contouring in matplotlib
fig, ax = plt.subplots(figsize=(12,8))
plt.axis('equal')

# matplotlib for contouring on triangle mesh
import matplotlib.tri as tri

femesh = tri.Triangulation(x,y,np.asarray(imat))
plt.triplot(femesh, lw=0.1, c='k')
plt.tricontourf(femesh, h, cmap='rainbow')
plt.colorbar(orientation='horizontal')
```

next, we create an animation writer to create a FFMPeg Movie File with a Framerate of 4 fps (this rather low vaue slows down the animation when showing the limited numer of time steps in the dac file).
The dac uses constant time steps too make this example easier.


```python
# https://matplotlib.org/examples/animation/moviewriter.html
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.animation as manimation

# initialize the writer
FFMpegWriter = manimation.writers['ffmpeg']
writer = FFMpegWriter(fps=4)
```

Finally, we step through the time steps of the dac, recreating the image each time. The `.grab_frame()` method adds the current image as the next frame to the movie. Upon termination, the writer saves the file.

```python
# render one image per time step
with writer.saving(fig, "writer_FEFLOW.mp4", fig.dpi):
    for i, t, dt in doc.getTimeSteps():
        print("TS # {}/{}: {}".format(i, doc.getNumberOfTimeSteps(), t))
        doc.loadTimeStep(i)        
        h = get_nodal_head_values(doc)
        
        # clear and replot figure
        ax.clear()
        ax.triplot(femesh, lw=0.1, c='k')
        ax.tricontourf(femesh, h, cmap='rainbow')
        
        # add fiure as frame to animation
        writer.grab_frame()
```
